```python
import torch
import gradio as gr
from diffusers import StableDiffusionPipeline

# Using fast SD-Turbo model
model_id = "stabilityai/sd-turbo"

# Loading model in half precision for speed
pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda")

# Enable memory efficient attention
pipe.enable_xformers_memory_efficient_attention()

# Image generation function
def generate_image(prompt):
    image = pipe(
        prompt=prompt,
        num_inference_steps=4,   # very low steps = very fast
        guidance_scale=0.0       # recommended for SD-Turbo
    ).images[0]

    return image

# Gradio UI
interface = gr.Interface(
    fn=generate_image,
    inputs=gr.Textbox(label="Enter your prompt"),
    outputs=gr.Image(label="Generated Image"),
    title="⚡ Fast AI Image Generator (Stable Diffusion Turbo)",
    description="Enter a prompt and generate images instantly!"
)

interface.launch()
```
